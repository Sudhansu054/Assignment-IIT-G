```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits import mplot3d
import matplotlib as mpl
```

```python
MeanVec = [0,0]
CovMatrix = np.matrix('1 0.6; 0.6 2')
```

```python
mu_x=MeanVec[0]
mu_y=MeanVec[1]
sigma_x=np.sqrt(CovMatrix[0,0])
sigma_y=np.sqrt(CovMatrix[1,1])
```

```python
x = np.random.normal(mu_x, sigma_x, 1000)
y = np.random.normal(mu_y, sigma_y, 1000)
```

```python
count, bins, ignored = plt.hist(x, 30, density=True)
plt.plot(bins, 1/(sigma_x * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu_x)**2 / (2 * sigma_x**2) ),
         linewidth=2, color='r')
```

```python
count, bins, ignored = plt.hist(y, 30, density=True)
plt.plot(bins, 1/(sigma_y * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu_y)**2 / (2 * sigma_y**2) ),
         linewidth=2, color='r')
```

```python
mv_x, mv_y = np.random.multivariate_normal(MeanVec,CovMatrix, 1000).T
```

```python
plt.plot(mv_x, mv_y, '*')
plt.axis('equal')
plt.show()
```

```python
L = np.linalg.cholesky(CovMatrix)
L
```

```python
t = np.linspace(0,2*np.pi,100)
C = [np.cos(t) , np.sin(t)]
C_x =np.cos(t) 
C_y =np.sin(t)
plt.plot(C_x,C_y, '-')
plt.show()
```

```python
E1 = 1*L*C
E2 = 2*L*C
E3 = 3*L*C
```
